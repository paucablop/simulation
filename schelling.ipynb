```python
from schelling.objects import Agent, Grid
from schelling.plot import plot_agents
from schelling.utils.generators import generate_grid, generate_agents_for_grid
```

```python
grid = generate_grid(size=100, threshold=0.2, occupancy=0.9, color_ratio=0.7)
grid.calculate_satisfactions()
plot_agents(grid.matrix)
```

```python
for i in range(100):
    grid.move_agents()
    grid.calculate_satisfactions()
    if i % 10 == 0:
        plot_agents(grid.matrix)
```

```python
for agent in grid.agents:
    print(agent.satisfaction) if agent.location == 6 else None
```
